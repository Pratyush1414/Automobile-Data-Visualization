```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
auto=pd.read_csv("../input/automobile/Automobile.csv")
```

```python
auto.head()
```

***Univariate distribution***

```python
sns.distplot(auto["normalized_losses"])
plt.show()
```

```python
sns.distplot(auto["highway_mpg"], kde= False, rug= True,color="red")
plt.show()
```

***Bivariate variation***

```python
sns.jointplot(auto["engine_size"], auto["horsepower"])
```

```python
sns.jointplot(auto["engine_size"], auto["city_mpg"],kind="hist", color="violet")
```

```python
sns.jointplot(auto["engine_size"],auto["horsepower"], kind="reg", color="orange")
```

**Multi-variate distribution**

```python
sns.pairplot(auto[["engine_size","horsepower","city_mpg","highway_mpg"]])
```

***2D Categorical against continuous variables**

```python
auto.head()
```

```python
sns.stripplot(auto["fuel_type"], auto["horsepower"], jitter= True)
```

```python
sns.swarmplot(auto["fuel_type"], auto["horsepower"])
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(auto["engine_location"], auto["horsepower"], hue=auto["fuel_type"])
```

```python
auto.head()
```

```python
sns.barplot(auto["body_style"], auto["horsepower"], hue=auto["engine_location"])
```

```python
sns.countplot(auto["body_style"])
```

```python
plt.figure(figsize=(10,10))
sns.pointplot(auto["make"], auto["horsepower"], hue= auto["number_of_doors"])
plt.xticks(rotation="vertical")
```

```python
sns.factorplot(data= auto, x="body_style", y="horsepower", col="fuel_type",kind="bar")
```

```python
sns.lmplot("horsepower","peak_rpm", hue="fuel_type",data=auto)
```
